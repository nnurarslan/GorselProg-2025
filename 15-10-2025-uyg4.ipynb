```python
secim = ""
while secim != "4":
    print("1 - toplama")
    print("2 - fark")
    print("3 - carpma")
    print("4 - cıkıs")
    secim = input("Seçiminizi yapın : ")
    
    if secim == "1":
        x = int(input("Birinci sayi : "))
        y = int(input("İkinci sayi : "))
        print(f"Sonuc {x+y}")
    elif secim == "2":
        x = int(input("Birinci sayi : "))
        y = int(input("İkinci sayi : "))
        print(f"Sonuc {x-y}")
    elif secim == "3":
        x = int(input("Birinci sayi : "))
        y = int(input("İkinci sayi : "))
        print(f"Sonuc {x*y}")
    elif secim == "4":
        print("Çıkış yapılıyor")
    else:
        print("Geçersiz seçim yaptınız")
        
        
```

```python
Kullanıcadan kaç tane not girmek istediğini soralım
for döngüsü ile notları tek tek alıp listeye atalım
durumlar listesi oluşturalım
notlar listesini for döngüsü ile gezelim 50 den büyük notların karşılığına
durumlar listesine "Geçti" ifadesini ekleyelim. Tersi durum "Kaldı" ekleyelim

En sonda notlar ve durum karşılıklarını yazdıralım
Örnek çıktı
Öğrenci 1 - Not : 55, Durum : Geçti
Öğrenci 2 - Not : 45, Durum : Kaldı
```

```python
not_sayisi = int(input("Kaç tane not gireceksiniz : "))
notlar = []
durumlar = []
for i in range(not_sayisi):
    notu = int(input(f"{i+1}. notu girin: "))
    notlar.append(notu)

for notu in notlar:
    if notu >= 50:
        durumlar.append("Geçti")
    else:
        durumlar.append("Kaldı")

for i in range(len(notlar)):
    print(f"Öğrenci {i+1} - Not: {notlar[i]}, Durum: {durumlar[i]}")
```
