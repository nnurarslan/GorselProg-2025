```python
MENU ADINDA BİR DICTIONARY OLUSTURUN
HER YEMEK ADI YEMEĞİN FİYATIYLA İLİŞKİLENDİRİLİR
KULLANICIDAN YEMEK ADI GİRMESİNİ İSTEYİN
YEMEK FİYATINI KONTROL EDİN
EĞER YEMEK SÖZLÜKTE VARSA
YEMEĞİN FİYATINA GÖRE UYGUN FİYATLI, ORTALAMA FİYAT VE PAHALI FİYAT DİYE YAZDIRIN
ÖRNEĞİN <50 UYGUN FİYAT
50-170 ARASI İSE ORTALAMA
170 DEN BÜYÜKSE PAHALI YAZSIN
EĞER YEMEK YOKSA KULLANICIYA MENÜDE BÖYLE BİR YEMEK YOK UYARISI VERSİN


```

```python
menu = {
    "pizza":275,
    "makarna":150,
    "hamburger":280,
    "simit":20
}
yemek_adi = input("Menüden yemek adı seçiniz")
if yemek_adi in menu:
    fiyat = menu[yemek_adi]
    if fiyat<50:
        durum = "uygun fiyatlı"
    elif 50 <= fiyat <= 170:
        durum ="orta fiyatlı"
    else:
        durum = "pahalı"
    print(f"Seçtiğiniz yemek : {yemek_adi}")
    print(f"Fiyat: {fiyat} TL")
    print(f"Durum : {durum}")
else:
    print("Menüde bu yemek yoktur")
```
