```python
class BankaHesabi:
    def __init__(self, hesap_sahibi, bakiye = 0):
        self.hesap_sahibi = hesap_sahibi
        self.__bakiye = bakiye #private variable
        
    def bakiye_goruntule(self):
        return self.__bakiye
    
    def para_yatir(self,miktar):
        self.__bakiye += miktar
    
    
        
```

```python
hesap = BankaHesabi("Ali",200)
print(hesap.hesap_sahibi)
print(hesap.__bakiye)
```

```python
hesap.bakiye_goruntule()
```

```python
hesap.para_yatir(500)
```

```python
class Calisan:
    def __init__(self, isim, maas):
        self.isim = isim
        self.maas = maas
    def bilgi_ver(self):
        print(f"İsim: {self.isim}, Maas : {self.maas}")

class Muhendis(Calisan):
    def __init__(self,isim,maas,uzmanlik_alani):
        self.uzmanlik_alani = uzmanlik_alani
        super().__init__(isim,maas)
        
    def bilgi_ver(self):
        super().bilgi_ver()
        print(f"Uzmanlik alani : {self.uzmanlik_alani}")

class Yonetici(Calisan):
    def __init__(self,isim,maas,ekip_sayisi):
        super().__init__(isim,maas)
        self.ekip_sayisi = ekip_sayisi
    
    def bilgi_ver(self):
        super().bilgi_ver()
        print(f"Ekip sayisi: {self.ekip_sayisi}")
        
        
```

```python
muhendis = Muhendis("Ahmet",70000,"Yazılım")
yonetici = Yonetici("Mehmet",100000,5)
```

```python
muhendis.bilgi_ver()
```

```python
yonetici.bilgi_ver()
```

```python
class Product:
    def __init__(self,name,price,stock):
        self.name = name
        self.price = price
        self.__stock = stock
    def check_stock(self,quantity):
        return self.__stock >= quantity
    def update_stock(self,quantity):
        if quantity <= self.__stock:
            self.__stock -= quantity
            return True
        return False
    def get_stock(self):
        return self.__stock
    
```

```python
class Customer:
    def __init__(self, name, email):
        self.name = name
        self.email = email
        self.__orders = []
    def add_order(self,order):
        self.__orders.append(order)
    def get_orders(self):
        return self.__orders
    
```

```python
class Order:
    def __init__(self, customer, product, quantity):
        self.customer=customer
        self.product=product
        self.quantity = quantity
        self.is_processed = False
        self.total_price = self.__calculate_total()
    def __calculate_total(self):
        return self.quantity * self.product.price
    def process_order(self):
        if self.product.check_stock(self.quantity):
            self.product.update_stock(self.quantity)
            self.is_processed = True
            self.customer.add_order(self)
            return True
        else:
            print("yetersiz stok")
            return False
        
```

```python
def display_order_history(customer):
    orders = customer.get_orders()
    if not orders:
        print(f"{customer.name} için sipariş geçmişi yok")
    else:
        print(f"{customer.name} için sipariş geçmişi:")
    for order in orders:
        status = "İşlendi" if order.is_processed else "Beklemede"
        print(f"Ürün:{order.product.name}, miktar: {order.quantity},{order.total_price},{status}")
```

```python
urun1 = Product("Laptop", 70000, 10)
urun2 = Product("Telefon", 50000,5)

musteri1 = Customer("Ali", "ali@gmail.com")
musteri2= Customer("Ayşe","a@gmail.com")

siparis1 = Order(musteri1, urun1, 2)
siparis2 = Order(musteri2, urun2, 4)
siparis3 = Order(musteri1, urun2, 2)

print("Sipariş işleme")
siparis1.process_order()
siparis2.process_order()
siparis3.process_order()

print("\nSipariş Geçmişi")
display_order_history(musteri1)
display_order_history(musteri2)
```
